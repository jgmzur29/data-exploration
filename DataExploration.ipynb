*DROPPING IRRELEVANT COLUMNS*

```python
df = df.drop(['Popularity'], axis=1)
df.head(5)
```

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

```python
df = df.drop(['Driven_Wheels'], axis=1)
df.head(10)
```

```python
df = df.rename(columns={"city mpg": "MPG"})
df.head(10)
```

```python
df = df.rename(columns={"Transmission Type": "Transmission", "Engine HP":"HP", "Engine Cylinders": "Cylinders"})
df.head(10)
```

*DROPPING DUPLICATES*

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("duplicated rows: ", duplicate_rows_df.shape)
```

```python
df.count()
```

So it's possible to observe that there are 11914 rows and we are removing 828 rows duplicated registers of data. df

```python
df = df.drop_duplicates()
df.head(10)
```

```python
df.count()
```

*DROPPING THE NULL/MISSING VALUES*

```python
print(df.isnull().sum())
```

we found 99 missing/null values. These values are the differences between the rows HP, Cylinders and the other ones. 

```python
df = df.dropna() #Dropping the missing values
```

```python
df.count()
```

Now we have some simetry at the rows of the data set

```python
print(df.isnull().sum())
```

*DETECTING OUTLIERS*

```python
sns.boxplot(x=df['MSRP'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

This last data its important to know the outliers and remove them from the data, due this data can affect or analysis or the models.

```python
df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
10988-9278
```

As seen above there were around 1710 rows outliers. Not all can be remove and sometimes the outliers also tell us something about the DS. <br>
